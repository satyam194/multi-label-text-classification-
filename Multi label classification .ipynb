```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from gensim import utils
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df =pd.read_csv(r"/content/Multi Label Text classification/Train.csv")
df.head()
```

```python
y=df.iloc[:, 6:]
y
```

```python
y=y.values.tolist()
```

```python
x=df["ABSTRACT"]
x
```

```python
'''from gensim import utils
import gensim.parsing.preprocessing as gsp

filters = [
          gsp.strip_punctuation,
           gsp.strip_multiple_whitespaces,
           gsp.strip_numeric,
           gsp.remove_stopwords,
           gsp.strip_short,
           gsp.stem_text
          ]
def clean_text(s):
  s=s.lower()
  s=utils.to_unicode
  for f in filters:
    s=f(s)
  return s'''

vectorizer =TfidfVectorizer(stop_words='english',max_features=10000, lowercase=True)
X=vectorizer.fit_transform(x)
```

```python
vectorizer.get_feature_names_out()
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,train_size=0.8)
```

```python
from sklearn.svm import LinearSVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.multiclass import OneVsOneClassifier
from sklearn.ensemble import RandomForestClassifier

clf=OneVsRestClassifier(LinearSVC()).fit(X_train,y_train)
#clf=RandomForestClassifier(max_depth=2, random_state=0)
#clf.fit(X, y)
```

```python
y_pred=clf.predict(X_test)
```

```python
def j_score(y_true, y_pred):
  jaccard = np.minimum(y_true, y_pred).sum(axis = 1)/np.maximum(y_true, y_pred).sum(axis = 1)
  return jaccard.mean()*100


def print_score(y_pred):
  print('Jacard score: {}'.format(j_score(y_test, y_pred)))
  print('----')

print_score(y_pred)
```
